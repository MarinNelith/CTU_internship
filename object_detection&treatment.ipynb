```python
from ultralytics import YOLO
from pathlib import Path
import cv2 as cv
import torch
import numpy as np
import os

# Create the 'output' folder if it doesn't exist
if not os.path.exists('output'):
    os.makedirs('output')

# Start webcam
cap = cv.VideoCapture(0)
cap.set(3, 1920)  # Width of the capture
cap.set(4, 1080)  # Height of the capture
cap.set(5, 30)  # FPS (Frames Per Second)
cap.set(cv.CAP_PROP_AUTOFOCUS, 0)  # Disable autofocus
initial_focus = 75  # Initial focus value
cap.set(cv.CAP_PROP_FOCUS, initial_focus)  # Set initial focus

# Model
model = YOLO("yolov8l-seg.pt")
model.to('cuda')
model.conf = 0.25

# Object classes
classNames = ["person", "bicycle", "car", "motorbike", "aeroplane", "bus", "train", "truck", "boat",
              "traffic light", "fire hydrant", "stop sign", "parking meter", "bench", "bird", "cat",
              "dog", "horse", "sheep", "cow", "elephant", "bear", "zebra", "giraffe", "backpack", "umbrella",
              "handbag", "tie", "suitcase", "frisbee", "skis", "snowboard", "sports ball", "kite", "baseball bat",
              "baseball glove", "skateboard", "surfboard", "tennis racket", "bottle", "wine glass", "cup",
              "fork", "knife", "spoon", "bowl", "banana", "apple", "sandwich", "orange", "broccoli",
              "carrot", "hot dog", "pizza", "donut", "cake", "chair", "sofa", "pottedplant", "bed",
              "diningtable", "toilet", "tvmonitor", "laptop", "mouse", "remote", "keyboard", "cell phone",
              "microwave", "oven", "toaster", "sink", "refrigerator", "book", "clock", "vase", "scissors",
              "teddy bear", "hair drier", "toothbrush"
              ]

while True:
    success, img = cap.read()
    if not success:
        break

    res = model.predict(img)

    for r in res:
        img = np.copy(r.orig_img)
        img_name = Path(r.path).stem

        for ci, c in enumerate(r):
            label = c.names[c.boxes.cls.tolist().pop()]
            b_mask = np.zeros(img.shape[:2], np.uint8)

            contour = c.masks.xy.pop().astype(np.int32).reshape(-1, 1, 2)
            _ = cv.drawContours(b_mask, [contour], -1, (255, 255, 255), cv.FILLED)

            mask3ch = cv.cvtColor(b_mask, cv.COLOR_GRAY2BGR)
            isolated = cv.bitwise_and(mask3ch, img)

            x1, y1, x2, y2 = c.boxes.xyxy.cpu().numpy().squeeze().astype(np.int32)
            iso_crop = isolated[y1:y2, x1:x2]

            filename = f'output/out_{label}_{ci}.jpg'
            cv.imwrite(filename, iso_crop)

            cv.imshow(f'Isolated Object {ci}', iso_crop)

        cv.imshow('Captured Image', img)

    # Key controls for adjusting focus
    key = cv.waitKey(1)
    if key == ord('q'):
        break
    elif key == ord('f'):
        initial_focus += 5
        if initial_focus > 255:  # Assuming the max focus value is 255
            initial_focus = 255
        cap.set(cv.CAP_PROP_FOCUS, initial_focus)
        print(f"Focus set to: {initial_focus}")
    elif key == ord('d'):
        initial_focus -= 5
        if initial_focus < 0:  # Assuming the min focus value is 0
            initial_focus = 0
        cap.set(cv.CAP_PROP_FOCUS, initial_focus)
        print(f"Focus set to: {initial_focus}")

cap.release()
cv.destroyAllWindows()
```
